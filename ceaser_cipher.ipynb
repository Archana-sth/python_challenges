```python
#function that takes a string and a shift value and returns the encoded message using the Caesar cipher

def caesar_cipher(text, shift):
    new_string = ""
    for char in text:
        if char.isalpha():        # check is the character is alphabet letter
            start = ord('a') if char.islower() else ord('A')  # starting value is uppercase or lowercase
            shifted_char = (ord(char)-start + shift)% 26 + start   # shift character and wrap around using modulo 26
            new_string += chr(shifted_char)
        else:
            new_string += char    #keeps non-letter character unchanged
    
    
    return new_string

caesar_cipher(" Hello ! World", 3)
```

```python
#function to decode messages by shifting in the opposite direction
def caesar_decode(text, shift):
    decrypted_message = ""
    for char in text:
        if char.isalpha():        # check is the character is alphabet letter
            start = ord('a') if char.islower() else ord('A')  # starting value is uppercase or lowercase
            new_char = (ord(char)-start - shift)%26 + start
            decrypted_message += chr(new_char)
        else:
            decrypted_message += char

    return decrypted_message

caesar_decode(" Khoor ! Zruog", 3)
```
