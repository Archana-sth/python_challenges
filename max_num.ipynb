```python
# function that takes a list of numbers and returns the maximum value in the list

def find_max(number):
    max_value = number[0]  # assume the firse element is the max value

    for n in number[1:]:
        if n > max_value:
            max_value = n
    return max_value
        

find_max ([-5, 9, -2, -12])



```

```python
#error handling for empty lists
def find_max(number):
   # max_value = number[0]
    if number == []:
        return "list is empty."
    else:
         max_value = number[0]
         for n in number[1:]:
            if n > max_value:
                max_value = n
    return max_value

find_max ([])
```

```python
#Return both the max value and its index in the list
def find_max(number):
    max_value = number[0]  # assume the firse element is the max value
    
    for n in number:   
        if n > max_value:
            max_value = n
            index = number.index(max_value)

    return max_value, index
        

value, idx = find_max ([-5, 9, -2, -12])  #find_max returns a tuple containing max value and index,so we need to unpack tuple into two variable

print (f"Max value : {value}")
print (f"Max index : {idx}")
```

```python
#Rewrite it using a while loop instead of a for loop
def find_max(number):
    max_value = number[0] 
    index = 1

    while index < len(number):
        if number[index] > max_value:
            max_value = number[index]
        index += 1
    return max_value

find_max([3,2,5,7,8])
```
