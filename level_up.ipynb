```python
#open file
with open("python_challenge.txt","w") as file:
    file.write("the cat and the hat.")
#file.close()
```

```python
#read file
with open("python_challenge.txt","r")as file:
    text = file.read()
print(text)
```

```python
#split into words
words = text.split()
```

```python
#count frequency
def word_count(filename):
    freq={}
    for word in words:
        if word in freq:
            freq[word] += 1
        else:
            freq[word] = 1
    return freq
word_count("python_challenge.txt")
```
