```python
def is_prime(number):
    if number <= 1:             # 0 and 1 are not prime
        return False 
    elif number == 2 or number == 3:   # 2 and 3 are prime number
        return True             
    elif number % 2 == 0:       #even numbers are not prime
        return False
    elif number % 3 == 0:
        return False
    elif number/number == 1:
        return True
is_prime(15)
```

```python
# function that returns all prime numbers up to a given number
def is_prime(limit):
    prime = []   # to store prime number
    for n in range(2, limit+1):     # range doesnt gives last number so limit +1 is the actual limit number
        if n == 2 or n ==3:
            return True
        elif n % 2 == 0:       #even numbers are not prime
            return False
        elif n % 3 == 0:
            return False
        n +=1
    return prime
        
is_prime(40)
```
