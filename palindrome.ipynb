```python
'''Palindrome is sequence of character that reads the same forwards and backward.'''
```

```python
def is_palindrome (word):
    word = "".join(filter(str.isalnum,word)).lower()   # converting to lowercase and removing spaces and punctuation.
    left = 0
    right = len(word)-1    # in the word "mam" length of word is 3 so 3-1 is the index 2 which is m,last letter in mam
    if word[left] == word[right]:
        return True
    else:
        return False
    left += 1
    right -=1

is_palindrome ("A man a plan a canal Panama")
```

```python
#Adapt the function to check if a number is a palindrome

def is_palindrome_number (n):
    n = str(n)  #converts number to string
    if n == n[::-1]:
        return True
    else:
        return False
is_palindrome_number (12333)
```

```python
# Return a message instead of a Boolean (e.g., "Yes, it's a palindrome!")

def is_palindrome (word):
    word = "".join(filter(str.isalnum,word)).lower()
    left = 0
    right = len(word)-1
    if word[left] == word[right]:
        return "Yes, it's a palindrome!"
    else:
        return "No, it's not a palindrome !"
    left += 1
    right -= 1


is_palindrome ("ma")

```
