```python
# loop method for reserved string


def single_string(word):
    reversed_word = " "             # start from blank value
    for letter in word:             #returns each letter in word
        reversed_word = letter + reversed_word  # each letter returns infront of reversed word eg: hello : h, eh, leh, lleh, olleh
    return reversed_word
single_string("hello")
```

```python
# function that removes whitespace or punctuation
def single_string(word):
    reversed_word = " "         
    for letter in word:
        if letter.isalnum():   # keeps only letter and number 
            reversed_word = letter + reversed_word  
    return reversed_word
single_string("hello     !!!!!  ")
```
